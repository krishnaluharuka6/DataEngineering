```python
import pandas as pd

df = pd.read_csv("demo_data_100.csv")
```

### Finding the missing or non-missing values

use isna(), isnull(), notna(), notnull() function

returns boolean data output

NA => Not Available (None, Not a Number, Not a Time)

```python
#detect missing values
df.isna()
df.isnull()
```

```python
# detect non-missing values
df.notna()
df.notnull()
```

```python
# Any missing values in each column
df.isna().any()
```

```python
# count missing values in each column
df.isna().sum()
```

### Drop any column from DatFrame

- use axis=1 parameter
- By default, the data is not modified in original dataframe, use inplace = True parameter

```python
df.drop("Age")
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```
